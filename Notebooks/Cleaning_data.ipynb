# Here, I changed the value of every league in the dataset that has a count of 1-10 to 'other'. Now, we have a total of 113 values with other

```python
league_counts = data['new_league'].value_counts()

# Find the leagues with count <= 10 and replace them with 'other'
other_mask = league_counts <= 10
data['new_league'] = data['new_league'].apply(lambda x: 'Other' if other_mask.loc[x] else x)

# Print the updated value counts
print(data['new_league'].value_counts())
```

# Here, I decided to reduce the number of values in the position column by categorizing them as shown below;

```python
def map_position(position):
    if position in ['Centre-Forward', 'Left Winger', 'Right Winger', 'Second Striker', 'attack']:
        return 'Forward'
    elif position in ['Central Midfield', 'Attacking Midfield', 'Defensive Midfield', 'Right Midfield', 'Left Midfield']:
        return 'Midfield'
    elif position in ['Centre-Back', 'Left-Back', 'Right-Back', 'defence']:
        return 'Defence'
    elif position == 'Goalkeeper':
        return 'Goalkeeper'
    else:
        return position

data['position'] = data['position'].apply(map_position)
```
