```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

```

```python
# loading my  data
df = pd.read_csv('../data/clean/clean_data.csv')


```

# creating a filter for players that were transferred for a fee of 3million euros and above.

```python
data = df.loc[(df['fees_paid_for_player'] >= 3)]
```

```python
data
```

```python
# extracting the columns of interest
market_value = data['market_value']
fees_paid_for_player = data['fees_paid_for_player']

# compute the Pearson correlation coefficient and p-value
corr, p_value = pearsonr(market_value, fees_paid_for_player)

# print the correlation coefficient and p-value
print(f"Correlation Coefficient: {corr:.4f}")
print(f"P-value: {p_value:.4f}")

# perform the hypothesis test at the 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant correlation between market_value and fees_paid_for_player.")
else:
    print("Fail to reject the null hypothesis. There is no significant correlation between market_value and fees_paid_for_player.")
```
